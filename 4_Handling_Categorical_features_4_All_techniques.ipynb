# Target Guided Ordinal Encoding
1. Ordering the labels according to the target
2. Replace the labels by the joint probability of being 1 or 0

```python
#titanic_data
t_d = pd.read_csv("/content/drive/MyDrive/Feature Engineering/titanic_train.csv",usecols=['survived','cabin'])
```

```python
t_d['cabin'].fillna('Missing',inplace=True)   # filling the NaN values with 'Missing'
```

```python
#let's take only the first letter from rows of cabin feature
t_d['cabin'] = t_d.cabin.astype('str').str[0]
```

```python
t_d['cabin'].unique()
```

```python
t_d.groupby('cabin')['survived'].mean().sort_values().index
```

```python
ordinal_lables = t_d.groupby('cabin')['survived'].mean().sort_values().index
```

```python
ordinal_label_count = {k:i for i,k in enumerate(ordinal_lables,0)}
```

```python
t_d['cabin_ordinal_lables'] = t_d['cabin'].map(ordinal_label_count)
```

# Mean Encoding 
mean encoding is same as ordinal encoding the main difference is <b>we replace category with it's mean<b/>

```python
mean_ordinal_lables = t_d.groupby('cabin')['survived'].mean().to_dict()
```

```python
mean_ordinal_lables
```

```python
t_d['mean_ordinal_encode'] = t_d['cabin'].map(mean_ordinal_lables)
```

# Probability Ratio Encoding
1. Probability of Survived based on Cabin --- Categorical Feature
2. Probability of Not Survived ---> 1-pr(Survived)
3. pr(Survived)/pr(Not Survived)I
4. Dictonary to map cabin with probability
5. replace with the categorical feature

```python
df = pd.read_csv("/content/drive/MyDrive/Feature Engineering/titanic_train.csv",usecols=['survived','cabin'])
```

```python
df.head()
```

```python
df['cabin'].fillna('Missing',inplace=True)  #replacing the NaN values with 'Missing'
```

```python
df.head(20)
```

```python
df['cabin'] = df.cabin.astype('str').str[0]   #extracing the first letter of data present in cabin
```

```python
df.head()
```

```python
df.cabin.unique()  #all unique values in cabin
```

```python
prob_df = df.groupby(['cabin'])['survived'].mean()
```

```python
prob_df = pd.DataFrame(prob_df)  # based on the cabin what's the percentage of survival
```

```python
type(prob_df)
```

```python
prob_df.head()
```

```python
prob_df['Died'] = 1-prob_df['survived']
```

```python
prob_df.head()
```

```python
prob_df['probability_ratio'] = prob_df['survived']/prob_df['Died']
```

```python
prob_df.head()
```

```python
probability_ratio_encoded = prob_df['probability_ratio'].to_dict()
```

```python
df['Cabin_encoded'] = df.cabin.map(probability_ratio_encoded)
```
