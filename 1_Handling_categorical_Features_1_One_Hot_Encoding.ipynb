# One Hot Encoding - variables with many categories

```python
data = pd.read_csv(r"/content/drive/MyDrive/Feature Engineering/mercedesbenz.csv",usecols=['X1','X2','X3','X4','X5','X6'])
```

```python
data.head()
```

```python
data.info()
```

```python
#how many categories/lables in each variable has
for col in data.columns:
    print(col, ':' ,len(data[col].unique()),'lables')
```

```python
#how many varibles will be there after one hot encoding?
pd.get_dummies(data,drop_first=True).shape
```

we can see that from just 6 initial categories variables,we get 117 new variables

Instead...
we would make a binary variable for each of the 10 most frequent variables only.
This is equivalent to grouping all the other lables under new category, that in this case wiil be dropped.Thus, the 10 new dummy variables indicate if one of the 10 most frequent lables is present (1) or not (0) for a particular obsevation.

```python
data.X2.value_counts().sort_values(ascending=False).head(15)
```

```python
#values with top 10 count of occurence of values in X2 
top10 = [x for x in data.X2.value_counts().sort_values(ascending=False).head(10).index]
top10
```

```python
for label in top10:
    data[label]=np.where(data['X2']==label,1,0)
data[['X2']+top10].head(30)
```

```python
#getting whole set of dummy variables for all categorical variables
def one_hot_topx(df ,variable ,top_x_lables):
    
    for label in top_x_lables:
        df[variable+'_'+label] = np.where(df[variable]==label,1,0)
    
#reading data again
data = pd.read_csv(r"/content/drive/MyDrive/Feature Engineering/mercedesbenz.csv",usecols=['X1','X2','X3','X4','X5','X6'])

#encode X2 into top most frequent categories 
one_hot_topx(data,'X2',top10)
data.head()
```

```python
#10 most frequent categories of X1
top10 = [x for x in data['X1'].value_counts().sort_values(ascending=False).head(10).index]
top10
```

```python
#encode X1 into top most frequent categories 
one_hot_topx(data ,'X1',top10)
data.head(10)
```

```python
#10 most frequent categories of X3
top10 = [x for x in data['X3'].value_counts().sort_values(ascending=False).head(10).index]
top10
```

```python
#encode X1 into top most frequent categories
one_hot_topx(data,'X3',top10)
```

```python
data.head(10)
```

```python
#10 most frequent categories of X4
top10 = [x for x in data.X4.value_counts().sort_values(ascending=False).head(10).index]
top10
```

```python
#encode X4 into top most frequent categories
one_hot_topx(data,'X4',top10)
```

```python
#encode X5 into top most frequent categories
one_hot_topx(data,'X5',top10)
```
